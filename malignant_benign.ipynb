**Factor Analysis**

```python
# Scaling the Features
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

```python
# Installing the library
!pip install factor_analyzer
```

```python
from factor_analyzer import FactorAnalyzer
FA = FactorAnalyzer(n_factors=2, rotation="varimax", method="principal", is_corr_matrix=False)
```

```python
FA.fit(X_scaled)
```

```python
print("Eigenvalues:")
print(FA.get_eigenvalues()[0])
```



*   Eigenvalues is used to represent the total amount of variance that can be given by a principal component.
*   Theortically they can be positive or negative but ideally they explain variance which are always positive



```python
print("Communalities:")
print(FA.get_communalities())
```

```python
print("Specific Variances:")
print(FA.get_uniquenesses())
```

```python
print("Factor Loadings:")
print(FA.loadings_)
```

```python
# creating the scree plot
import matplotlib.pyplot as plt
plt.style.use("ggplot")

plt.plot(FA.get_eigenvalues()[0], marker='o')
plt.xlabel("Eigenvalue number")
plt.ylabel("Eigenvalue size")
plt.title("Scree Plot")
```

```python
# Creating new feature dataset
transformed_df = pd.DataFrame(FA.transform(X_scaled), columns=['RF1', 'RF2'])
```
